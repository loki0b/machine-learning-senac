**Processo de união com as demais tabelas disponibilizadas:**

1 - (Clientes, Entregas)

2 - (1, Pedidos)

3 - (2, Pagamentos)

4 - (3, Review)

5 - (4, Vendedor)

```python
merge_table = pd.merge(tabela_cliente, tabela_entrega_pedido, on=['customer_id'], how='inner')
merge_table = pd.merge(merge_table, tabela_pedido, on=['order_id'], how='inner')
merge_table = pd.merge(merge_table, tabela_pagamento, on=['order_id'], how='inner')
merge_table = pd.merge(merge_table, tabela_review, on=['order_id'], how='inner')
merge_table = pd.merge(merge_table, tabela_vendedor, on=['seller_id'], how='inner')

merge_table
```

```python
merge_table.columns.tolist
```

**Cópia do dataframe**

```python
merge_table_copy = merge_table.copy()
```

**Descartando algumas colunas a principio não necessárias** 

```python
merge_table_copy = merge_table.drop(columns=['customer_unique_id', 'customer_city', 'customer_state', 'order_status',
       'order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date', 'shipping_limit_date','review_creation_date',
       'review_answer_timestamp','seller_city',
       'seller_state', 'review_comment_title', 'review_comment_message'])
```

**Transformando em arrays numpy**

Para melhor manipulação - desempenho

```python
vetor_cliente = np.array(merge_table_copy.customer_id)
vetor_cepcliente = np.array(merge_table_copy.customer_zip_code_prefix)
vetor_pedido = np.array(merge_table_copy.order_id)
vetor_id_pedido = np.array(merge_table_copy.order_item_id)
vetor_produto = np.array(merge_table_copy.product_id)
vetor_vendedor = np.array(merge_table_copy.seller_id)
vetor_preco_produto = np.array(merge_table_copy.price)
vetor_frete = np.array(merge_table_copy.freight_value)
vetor_parcela = np.array(merge_table_copy.payment_sequential)
vetor_tipopagamento = np.array(merge_table_copy.payment_type)
vetor_pay = np.array(merge_table_copy.payment_installments)
vetor_valorfinal = np.array(merge_table_copy.payment_value)
vetor_review = np.array(merge_table_copy.review_id)
vetor_score = np.array(merge_table_copy.review_score)
vetor_cepvendedor = np.array(merge_table_copy.seller_zip_code_prefix)
```

**Novo DataFrame vazio**

```python
data_frame = pd.DataFrame()

data_frame
```

**Definindo as colunas do novo dataframe e atribuindo para cada coluna, seu respectivo vetor de dados registrado anteriormente**

```python
COLUNAS = ['Cliente',
'CEP_Cliente',
'Pedido',
'id_Pedido',
'Produto',
'Vendedor',
'Preco_produto',
'Frete',
'Parcela',
'Tipo_pagamento',
'Installments',
'Valor_total',
'ID_Review',
'CEP_Vendedor',
'Score'
]
```

```python
data_frame = pd.DataFrame(columns = COLUNAS)

data_frame.Cliente = vetor_cliente
data_frame.CEP_Cliente = vetor_cepcliente
data_frame.Pedido = vetor_pedido
data_frame.id_Pedido = vetor_id_pedido
data_frame.Produto = vetor_produto
data_frame.Vendedor = vetor_vendedor
data_frame.Preco_produto = vetor_preco_produto
data_frame.Frete = vetor_frete
data_frame.Parcela = vetor_parcela
data_frame.Tipo_pagamento = vetor_tipopagamento
data_frame.Installments = vetor_pay
data_frame.Valor_total = vetor_valorfinal
data_frame.ID_Review = vetor_review
data_frame.CEP_Vendedor = vetor_cepvendedor
data_frame.Score = vetor_score

data_frame
```

**Importando os métodos de árvore de decisão**

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

**Para simplificar os dados, removemos todos os elementos não numéricos para que o modelo seja capaz de realizar a execução**

```python
data_frame['Cliente'] = data_frame['Cliente'].str.replace(r'\D', '')
data_frame['Pedido'] = data_frame['Pedido'].str.replace(r'\D', '')
data_frame['Produto'] = data_frame['Produto'].str.replace(r'\D', '')
data_frame['Vendedor'] = data_frame['Vendedor'].str.replace(r'\D', '')
data_frame['ID_Review'] = data_frame['ID_Review'].str.replace(r'\D', '')

data_frame
```

```python
result_data_frame = pd.get_dummies(data_frame, columns=['Tipo_pagamento'])
result_data_frame
```

**Criando um backup**

```python
reserva = result_data_frame.copy()

reserva
```

**Eliminando clientes com mais de 3 estrelas de satisfação**

```python
reserva = reserva.drop(reserva[reserva.Score > 3].index)

reserva
```

**Processo de separação de treino/teste**

Proporções

70% treino

30% teste

```python
y = result_data_frame['Score']

X = result_data_frame.drop('Score', axis=1)

X_train, X_teste, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

**Número de amostras para cada processo**

```python
X_train.shape, X_teste.shape
```

**Número de targets para cada processo**

```python
y_train.shape, y_test.shape
```

**Criação do classificador**

```python
cls = DecisionTreeClassifier()
```

**Treinamento**

```python
cls = cls.fit(X_train, y_train)
```

**Vetor com as importancias de cada atributo para a determinação do review_score**

```python
result_data_frame.columns.to_list
```

```python
cls.feature_importances_
```

```python
from sklearn.tree import plot_tree, export_text
```

```python
plt.rcParams['figure.figsize'] = (40, 20)
```

```python
plt.figure()
plot_tree(cls, filled=False)
plt.show()
```

```python
data_frame.head()
```

```python
for feature, importancia in zip(result_data_frame.columns, cls.feature_importances_):
    print(f'{feature}:{importancia*100:.1f}%')
```

**Vetor com predições para checagem do aprendizado**

```python
result = cls.predict(X_teste)
result
```

**Representação das métricas de precisão e médias do modelo**

```python
print(metrics.classification_report(y_test, result))
```

**Precisão total**

```python
from sklearn.model_selection import cross_val_score

allScores = cross_val_score(cls, X_train, y_train, cv=10)
allScores.mean()
```
