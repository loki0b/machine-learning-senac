```python
import pandas as od
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple', 'Google', 'Amazon', 'Microsoft', 'Tencent', 'McDonalds', 'Visa', 'Facebook', 'Alibaba', 'Louis Vouitton']
bilhoes = [947.062, 819.573, 705.646, 611.460, 214.023, 196.526, 191.032, 186.421, 169.966, 124.273]
```

```python
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhes')
```

```python
plt.rcParams['figure.figsize'] = (15, 8)

plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhes')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhes', fontsize=25)
```

```python
plt.plot(empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhes', fontsize=25)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
```

```python
plt.plot(empresas, bilhoes, color = '#8844ee', lw = 10)
plt.title('Empresas mais ricas do mundo', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhes', fontsize=25)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
```

```python
plt.plot(empresas, bilhoes, color = '#8844ee', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhes', fontsize=25)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
```

```python
plt.plot(empresas, bilhoes, color = '#8844ee', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhes', fontsize=25)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.grid(True)
```
